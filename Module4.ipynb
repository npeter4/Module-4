```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

df_cleaned = pd.read_csv('2023-2024_NBA_Player_Stats_Cleaned.csv')

df_sorted = df_cleaned.sort_values(by='Tm', ascending=True)

df_unique_players = df_sorted.drop_duplicates(subset='Player', keep='first')

df_features = df_unique_players[['PTS', 'AST', 'TRB', 'FG%']]

inertia_values = []

for k in range(2, 31):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_features)
    inertia_values.append(kmeans.inertia_)

plt.figure(figsize=(15, 6))
plt.plot(range(2, 31), inertia_values, marker='o')
plt.title('K-Means Inertia (Elbow Method)')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
df_unique_players['Cluster'] = kmeans.fit_predict(df_features)
```

```python

cluster_0_players = df_unique_players[df_unique_players['Cluster'] == 0]
top_20_cluster_0 = cluster_0_players.head(20)
print(top_20_cluster_0[['Player', 'PTS', 'AST', 'TRB', 'FG%', 'Cluster']])
```
